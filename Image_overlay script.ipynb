```python
train_images=get_images_dr("hand_recognition/data/train")
test_images=get_images_dr("hand_recognition/data/validation")
```

```python
def aug_save(path1,path2,save_dir):
    log_arr=np.array(Image.open(path1))
    hand_arr=np.array(Image.open(path2))
    log_arr=ms.imresize(log_arr,(10,10,3))
    log_arr=ms.imresize(log_arr,hand_arr.shape)
    r=rand.choice(np.arange(0.6,0.8,.02))
    result=cv2.addWeighted(hand_arr,r,log_arr,r,0)
    p="/".join(path2.split("/")[3:])
    ms.imsave(save_dir+p,result)
    
```

```python
for i in range(len(train_images)):
    aug_save(logos[i],train_images[i],"augmented_train/")
    
```

```python
os.mkdir("augmented_test")
for dr in os.listdir("hand_recognition/data/validation"):
    os.mkdir("augmented_test/"+dr)
```

```python
logos_test=logos[len(test_images):]
for i in range(len(test_images)):
    aug_save(logos_test[i],test_images[i],"augmented_test/")
```
